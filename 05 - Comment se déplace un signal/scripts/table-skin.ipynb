```python
import pandas as pd
import os
```

```python

df = pd.DataFrame({
    "Material": [r"Copper (Cu)", r"Silver (Ag)", r"Aluminum (Al)", r"Gold (Au)", r"Titanium (Ti)"],
    r"\makecell{Skin Depth\\at 100MHz $(\unit{\mu m})$}": [6.52, 6.34, 8.19, 7.489, 103.1],
    r"\makecell{Skin Depth\\at 1GHz $(\unit{\mu m})$}": [2.06, 2.01, 2.59, 2.37, 32.6],
    r"\makecell{Skin Depth\\at 10GHz $(\unit{\mu m})$}": [0.652, 0.634, 0.819, 0.749, 10.3],
    r"\makecell{Skin Depth\\at 100GHz $(\unit{\mu m})$}": [3.66, 0.201, 0.259, 0.237, 3.26]
})

df
```

```python
# Output path (make sure directory exists)
output_path = "../tables/skin-depth.tex"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Generate LaTeX string
latex = df.to_latex(
    index=False,
    escape=False,
    header=True,
    column_format='l|c|c|c|c',
    float_format="%.2f",
    bold_rows=True,
    caption="Skin depth of different material (MW design p.173)", 
    label="tab:skin-depth",
)

# Inject \hline after the header
lines = latex.splitlines()
for i, line in enumerate(lines):
    if r'\midrule' in line:
        lines.insert(i + 1, r'\hline')
        break
latex = '\n'.join(lines)

# Write to file
with open(output_path, "w") as f:
    f.write(latex)
```
